```python
# Functions for different filters

def make_grayscale(img):
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return gray_img


def filter_img_sobelx(img):
    sobelx = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
    return sobelx

def normalize_img255(img):
    img = img/255.0
    return img

def normalize_img(img):
    img = img - np.min(img)
    img = img/np.max(img)
    return img

def sharpen_image(img):
    kernel = np.array([[0, -1, 0],
                       [-1, 5,-1],
                       [0, -1, 0]])
    image_sharp = cv2.filter2D(src=img, ddepth=-1, kernel=kernel)
    return image_sharp

def Gaussian_blur(img):
    kernel = np.array([[1/16, 2/16, 1/16],
                       [2/16, 4/16, 2/16],
                       [1/16, 2/16, 1/16]])
    image_blur = cv2.filter2D(src=img, ddepth=-1, kernel=kernel)
    return image_blur
```

```python
# Reading the data and applying filters on the data

images = []
grayimages = []
filteredimages = []
norrimg = []
modified_filtered = []
modified_color_filtered = []


for img in mportPaths:
    img = cv2.imread(img).astype(np.float32)
    img = cv2.resize(img, (SIZE, SIZE))
    gray_img = make_grayscale(img)
#     img = normalize_img255(img)
#     filtered_img = filter_img_sobelx(gray_img)
#     norr_img = normalize_img(gray_img)
    blur_image = Gaussian_blur(gray_img)
    sharpened_image = sharpen_image(blur_image)
    
    color_blur_img = Gaussian_blur(img)
    color_sharpened_image = sharpen_image(color_blur_img)

    # gathering data into list
    images.append(img)
    grayimages.append(gray_img)
#     filteredimages.append(filtered_img)
#     norrimg.append(norr_img)
    modified_filtered.append(sharpened_image)
    modified_color_filtered.append(color_sharpened_image)

```

```python
#Converting filtered data to array

nor_img = np.array(modified_filtered)

col_nor_img = np.array(modified_color_filtered)
```
